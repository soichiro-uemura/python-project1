このキーワードリストのcsvファイル"course_keyword_list.csv"と授業データのjsonファイル"catalog-2018.json"を使って、以下の関数count_keywordを作成してください。

関数count_keywordは、授業の開講所属学部名（理学部、工学部、法学部、など）を引数として受け取り、その学部の開講講義の題目について、キーワードリスト中の各キーワードを含む題目の数を集計して、上記と同じく、キーワードとそのキーワードを含む講義題目数の組みをコンマ区切りで講義題目数の降順でcsvファイル"output.csv"に出力します。なお、csvファイルの読み出し・書き出しにはencoding=’utf-8’を指定してください。

題目、学部名は、それぞれ、jsonファイル"catalog-2018.json"の"title_j"キーの値、"department_j"キーの値を使用してください。

なお、辞書の要素をキーの値で並び替えるのに以下の関数`sort_dicvalue`を使用してもよいです。関数`sort_dicvalue`では、引数で受け取った辞書について、その要素のキーの値で降順で要素を並び替え、その結果をキーと値のタプルで順に返します。関数`sort_dicvalue`の返り値は、以下のように`for`文で順に取り出すことができます。
```Python
def sort_dicvalue(mydic):
    for k,v in sorted(mydic.items(), key=lambda x:x[1], reverse=True):
        yield (k, v)
    
for k,v in sort_dicvalue(辞書に対応する変数):
    print (k,v)
```

````python
def count_keyword(dep):
    import csv
    A=[]
    with open('course_keyword_list.csv', 'r',encoding='utf-8') as f:
        dataReader = csv.reader(f)
        for row in dataReader:
            A.append(row[0])
    B=[0]*len(A)
    import json
    with open("catalog-2018.json", "r", encoding="utf-8") as g:
        j = json.load(g)
        for d in j:
            if dep in d.get('department_j',''):
                for i in range(len(A)):
                     if A[i] in d.get('title_j',''):
                        B[i] +=1
    C={A[k]:B[k] for k in range(len(A))}
    def sort_dicvalue(mydic):
        for l,m in sorted(mydic.items(), key=lambda x:x[1], reverse=True):
             yield (l, m)
    with open('output.csv', 'w',encoding="utf-8") as h:
        dataWriter = csv.writer(h, lineterminator='\n')
        for l,m in sort_dicvalue(C):
            dataWriter.writerow((l,m))
````
